```python
import numpy as np
import polars as pl
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

from jormund import *

# Define parameters
order = (1, 0, 1)   # Example ARIMA order (p, d, q)
n = 1000              # Number of observations

# Set up ARIMA model and simulate
arima_process = sm.tsa.ArmaProcess.from_coeffs([0.5], [-0.4])  # Define your AR and MA coefficients here
simulated_data = pl.DataFrame(arima_process.generate_sample(nsample=n))
```

```python
plt.boxplot(simulated_data)
plt.show()
```

```python
plt.hist(simulated_data)
plt.show()
```

```python
plt.plot(simulated_data)
plt.show()
```

```python
plot_acf(simulated_data["column_0"], lags=100)
plt.show()
```
