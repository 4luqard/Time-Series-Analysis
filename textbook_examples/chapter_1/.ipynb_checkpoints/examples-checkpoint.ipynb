```python
import matplotlib.pyplot as plt
import polars as pl
import pandas as pd
import numpy as np
import tqdm

pl.Config.set_tbl_cols(None)
```

# Exercise 1:

```python
larain = pl.read_csv('larain.csv')
years = range(1878, 1878 + len(larain))
larain
```

```python
plt.figure(figsize=(12, 4))

plt.plot(years, larain['larain'], marker='o', linestyle='-', color='teal')

plt.xlabel('Year', fontsize=12)
plt.ylabel('Inches', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.scatter(larain.shift(1), larain, marker='o', linestyle='-', color='teal')

plt.xlabel('Previous Year Inches', fontsize=12)
plt.ylabel('Inches', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
diff = larain - larain.shift(1)
diff.head(3)
```

```python
plt.figure(figsize=(12, 4))

plt.plot(years, diff, marker='o', linestyle='-', color='purple')

plt.xlabel('Previous Year Inches', fontsize=12)
plt.ylabel('Inches', fontsize=12)

plt.tight_layout()
plt.show()
```

# Exercise 3:

```python
mean = 0
std = 1
random_numbers = np.random.normal(mean, std, 48)
random_numbers
```

```python
plt.figure(figsize=(12, 4))

plt.plot(range(len(random_numbers)), random_numbers, marker='o', linestyle='-', color='purple')

plt.xlabel('Batch', fontsize=12)
plt.ylabel('values', fontsize=12)

plt.tight_layout()
plt.show()
```

# Exercise 4:

```python
degrees_of_freedom = 2
chi_square = np.random.chisquare(degrees_of_freedom, 48)
chi_square
```

```python
plt.figure(figsize=(12, 4))

plt.plot(range(len(chi_square)), chi_square, marker='o', linestyle='-', color='purple')

plt.xlabel('Batch', fontsize=12)
plt.ylabel('values', fontsize=12)

plt.tight_layout()
plt.show()
```
